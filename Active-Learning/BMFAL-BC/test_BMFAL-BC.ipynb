```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

with open("__results__/Poisson2/res_GreedySubmod_trail1.pickle", "rb") as f:
    results = pickle.load(f)

# Compute the mean RMSE
mean_rmse = np.mean(results["hist_test_nRmse"], axis=(1, 2))

plt.figure(figsize=(10, 7))
plt.plot(results["hist_cost"])
plt.title('Cost History')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.show()

plt.figure(figsize=(10, 7))
plt.plot(mean_rmse)
plt.title('Mean Test RMSE History')
plt.xlabel('Iteration')
plt.ylabel('Mean Test RMSE')
plt.show()

# Compute the mean RMSE for the ground
mean_rmse_ground = np.mean(results["hist_test_nRmse_ground"], axis=(1, 2))

plt.figure(figsize=(10, 7))
plt.plot(mean_rmse_ground)
plt.title('Mean Test Ground RMSE History')
plt.xlabel('Iteration')
plt.ylabel('Mean Test Ground RMSE')
plt.show()
```
