```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from csv files
s0r0 = np.loadtxt("k5f1_s0r0.csv")
s0r1 = np.loadtxt("k5f1_s0r1.csv")
s1r0 = np.loadtxt("k5f1_s1r0.csv")
s1r1 = np.loadtxt("k5f1_s1r1.csv")

plt.figure(figsize=(10, 7))

plt.plot(s0r0, label='s0r0')
plt.plot(s0r1, label='s0r1')
plt.plot(s1r0, label='s1r0')
plt.plot(s1r1, label='s1r1')

plt.legend()

plt.title('RMSE over epochs')
plt.xlabel('Epochs')
plt.ylabel('RMSE')

plt.show()
```
