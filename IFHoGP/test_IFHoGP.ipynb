```python
import torch

model = torch.load('__res_heat__/dmf/rank5/fold1/dicts/model.pt')
print(model['mf_hidden_models.0.Aw'].numpy())
```

```python
import matplotlib.pyplot as plt

import numpy as np

weights = model['mf_hidden_models.0.Aw'].numpy()

weights = np.log(np.abs(weights))

# Normalize the data to the range [0, 1]
weights = (weights - weights.min()) / (weights.max() - weights.min())

plt.imshow(weights, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
bias = model['mf_hidden_models.0.Ab'].numpy()

# Plot the bias
plt.figure(figsize=(10, 5)) 
plt.plot(bias)  
plt.show()
```
