```python
import matplotlib.pyplot as plt
import pickle

with open("__res__/Burgers/InfFidGPT/base5/fold0/best_pred_list.pickle", "rb") as f:
    data = pickle.load(f)

pred_list = data['pred_list']

for i in range(5):
    plt.figure(figsize=(10, 7))
    first_layer = pred_list[16][i]
    plt.imshow(first_layer, cmap='gray')
    plt.colorbar()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a grid for 3D plot
x = np.arange(16)
y = np.arange(16)
x, y = np.meshgrid(x, y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

for i, z_level in enumerate(pred_list[16][:5]):
    z = np.full((16,16), i*20)  # Create a level for each image, here we separate them by 20 units
    ax.plot_surface(x, y, z, rstride=1, cstride=1, facecolors=plt.cm.viridis(pred_list[16][i]))

plt.show()
```
